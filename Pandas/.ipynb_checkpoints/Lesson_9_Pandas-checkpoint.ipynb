### Объединение датафреймов

__merge__

```python
d1 = pd.DataFrame(
    {"key": ["a", "b", "c", "c", "d", "e"],
    "num1":range(6)})
d2 = pd.DataFrame(
    {"key":["b", "c", "e", "f"],
    "num2":range(4)})
```

```python
d3 = pd.DataFrame(
    {"key1": ["a", "b", "c", "c", "d", "e"],
    "num1":range(6)})
d4 = pd.DataFrame(
    {"key2":["b", "c", "e", "f"],
    "num2":range(4)})
```

```python
d1
```

```python
d2
```

```python
# Слияние датафреймов по key:
pd.merge(d1, d2)
# Получаются все значения, которые есть и в первом, и во втором датафрейме
```

```python
# Эта запись равносильна следующей записи:
pd.merge(d1, d2, on="key")
```

```python
d3
```

```python
d4
```

```python
# Слияние двух датафреймов по общим ключам (key) с помощью left_on, right_on
pd.merge(d3, d4, left_on="key1", right_on="key2")
```

```python
# Хотим, чтобы при слиянии попали все значения и все ключи, которые есть в двух датафреймах:
pd.merge(d1, d2, how="outer")
```

```python
# Хотим, чтобы при слиянии попали все значения из левого датафрейма, а из правого - 
# только те, которые есть в левом
pd.merge(d1, d2, how="left")
```

```python
# Наоборот:
pd.merge(d1, d2, how="right")
```

```python
pd.merge(d1, d2, how="inner")  # Только пересечения
```

```python
df1=pd.DataFrame(
    {"key":["a","b", "c", "c", "d", "e"],
    "num1":range(6),
    "count":["one", "three", "two", "one", "one", "two"]})
df2=pd.DataFrame(
    {"key":["b", "c", "e", "f"],
    "num2":range(4),
    "count":["one", "two", "two", "two"]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on=["key", "count"], how="outer")
```

```python
# При одинаковых именах столбцов, они дублируются с индексом
pd.merge(df1, df2, on="key", how="outer")
# Слияние по key, значит будет один столбец key. Count - одинаковый столбец, их будет два
```

```python
pd.merge(df1, df2, on="key", suffixes=("_data1", "_data2"))
```

```python
# Если разные названия столбцов:
```

```python
df1=pd.DataFrame(
    {"letter":['a', 'a', 'b', 'b', 'a', 'c'],
    "num":range(6)})
df2=pd.DataFrame(
    {"value":[3,5,7]},
    index=['a', 'b', 'c'])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, left_on="letter", right_index=True)
```

```python
right=pd.DataFrame(
    [[1,2],[3,4],[5,6]],
    index=["a","c","d"],
    columns=["Tom","Tim"])
left=pd.DataFrame(
    [[7,8],[9,10],[11,12],[13,14]],
    index=["a","b","e","f"],
    columns=["Sam","Kim"])
```

```python
# Полное слияние двух таблиц
pd.merge(right,left, 
         right_index=True, 
         left_index=True, 
         how="outer")
# Общие только индексы, поэтому right_index=True, left_index=True
```

__join__  
если к левому датафрейму добавляем правый, пишем: left.join(right): при этом на выходе будет сначала левый, потом правый
по умолчанию: inner

```python
left.join(right)
```

```python
left.join(right, how='outer')
```

```python
data=pd.DataFrame([[1,3],[5,7],[9,11]],            
                  index=["a","b","f"],      
                  columns=["Alex","Keta"])
left.join([right,data])
```

__concatenate в Numpy__

```python
seq= np.arange(20).reshape((4, 5))
```

```python
np.concatenate([seq, seq], axis=1)
# Слияние по вертикали - увеличивается количество столбцов
```

```python
np.concatenate([seq, seq], axis=0)
# Слияние по горизонтали - увеличивается количество строк
```

__concatenate в Pandas__

```python
data1 = pd.Series(
    [0, 1], index=['a', 'b'])
data2 = pd.Series(
    [2, 3, 4], index=['c','d','e'])
data3 = pd.Series(
    [5, 6], index=['f', 'g'])
```

```python
data1, data2, data3
```

```python
pd.concat([data1, data2, data3])
```

```python
pd.concat([data1, data2, data3], axis=1)
# Series аавтоматически превратится в столбец после добавления оси axis
```

```python
data4= pd.Series([10,11,12], 
                 index=['a','b',"c"])
data4
```

```python
pd.concat([data1, data4], axis=1, join='inner')
```

```python
# При использовании concat для данных можно добавлять ключи
pd.concat([data1, data2, data3], keys=['one', 'two', 'three'])
```

```python
pd.concat([data1, data2, data3], axis=1, keys=['one', 'two', 'three'])
```

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['one', 'two'])
df2 = pd.DataFrame(
    10+np.arange(4).reshape(2,2),
    index=['a', 'c'],
    columns=['three', 'four'])
```

```python
df1
```

```python
df2
```

```python
# Создаем вертикальные разделы:
pd.concat([df1, df2], axis=1, keys=['s1', 's2'])
```

```python
pd.concat([df1, df2], axis=1, 
          keys=['s1', 's2'],
          sort=False)
```
