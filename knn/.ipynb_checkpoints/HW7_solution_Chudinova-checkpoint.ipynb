**Создаем вектор признаков и целевую переменную:**

```python
X = iris.drop(['target'], axis = 1)
y = iris['target']
```

**Делим выборку на обучающую (train) и тестовую (test):**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

**Feature scaling: нормализуем признаки:**

```python
cols = X_train.columns
cols
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns = [cols])
X_test = pd.DataFrame(X_test, columns = [cols])
```

```python
X_train.head()
```

**Обучаем при k = 3:**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

**Предсказываем по тестовой выборке:**

```python
y_pred = knn.predict(X_test)
y_pred
```

**Предсказываем цветок отдельно каждого из трех видов:**  
0 - *Iris setosa*, 1 - *ris versicolor*, 2 - *Iris virginica*

```python
knn.predict_proba(X_test)[:,0]
```

Предсказание вида *Iris setosa* произведено с вероятностью 100%. Это согласуется с анализом распределения значений признаков по трем видам ириса: именно вид *Iris setosa* можно было полностью отличить от двух других видов.

```python
knn.predict_proba(X_test)[:,1]
```

```python
knn.predict_proba(X_test)[:,2]
```

Предсказание видов *Iris versicolor* и *Iris virginica* проведено с вероятностями 100%, 66,7% и 33,3% в связи с частичным перекрытием значений всех четырех призаков для данных двух видов ирисов.

**Accuracy:**

Поскольку классы целевой переменной по количеству значений равны, оценим метрику accuracy.

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
y_pred_train = knn.predict(X_train)
print('Training set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

```python
print('Training set score: {:.4f}'.format(knn.score(X_train, y_train)))
print('Test set score: {:.4f}'.format(knn.score(X_test, y_test)))
```

Точности тестового и обучающего наборов очень сходны, соответственно модель не переучена и не недоучена. Полученная модель имеет достаточно высокую точность равную 0.9778, то есть доля верных ответов будет составлять 97.98%.

**Нулевая точность модели:**

```python
y_test.value_counts()
```

```python
null_accuracy = (y_test.value_counts()[1]/y_test.value_counts().sum())
print('Null accuracy score: {0:0.4f}'. format(null_accuracy))
```

Точность модели, равная 0.9778, значительно превышает нулевую точность 0.4.

**Оцениваем модели с разным значением соседей k:**

```python
print("Model accuracy score with k:")
for idx, x in enumerate(range(1, 21)):
    knn_x = KNeighborsClassifier(n_neighbors=x)
    knn_x.fit(X_train, y_train)
    y_pred_x = knn_x.predict(X_test)
    print("k =", idx + 1, "  ", round(accuracy_score(y_test, y_pred_x), 4))
```

Точность модели увеличивается при увеличении k до значения 3. Со значения k=3 до k=16 точность модели не изменяется; с последующим увеличением k точность модели уменьшается. Оптимальное значение k - минимальное значение k при максимальной точности: k = 3. Следовательно, изначально значение k было выбрано правильно.

**Анализ ошибок модели с помощью Confusion matrix:**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm_matrix = pd.DataFrame(cm, columns=['Predicted I.setosa', 'Predicted I.versicolor', 'Predicted I.virginica'], 
                        index=['True I.setosa', 'True I.versicolor', 'True I.virginica'])
cm_matrix
```

```python
true_label = np.diag(cm).sum()
false_label = cm.sum() - true_label
print("True label =", true_label)
print("False label =", false_label)
```

```python
plt.figure(figsize=(7,4))
sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
plt.yticks(rotation=0)
plt.show()
```

Поскольку целевая переменная не является бинарной, а содержит 3 класса, то с помощью confusion matrix мы рассчитываем true label (верные значений) *vs.* false label (неверные значения). Для нашей модели с k = 3, количество верных значений равно 44, количество ошибочных значений равно 1.

**Можно посмотреть метрики для модели с помощью classification report:**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, digits = 4))
```

Все три метрики - прецизионность, полнота и их среднее гармоническое - равны 100% для класса 0 (*Iris setosa*). Для класса 1 (*Iris versicolor*) модель предсказывает верные объекты (прецизионность) в 100% случаев, а целевые объекты из всех объектов (полнота) в 94.44% случаев. Для класса 2 (*Iris virginica*) модель предсказывает верные объекты в 91.67% случаев, а целевые объекты из всех объектов в 100% случаев. Средняя гармоническая прецизионности и полноты самая низкая - для *Iris virginica*.
