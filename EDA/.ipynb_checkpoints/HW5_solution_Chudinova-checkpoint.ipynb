### HW5 Autism_Data

Расстройство аутистического спектра (РАС) — это нарушение развития нервной системы, связанное со значительными затратами на здравоохранение, и ранняя диагностика может значительно сократить их. К сожалению, время ожидания диагноза РАС длительное, а процедуры неэффективны с точки зрения затрат. Экономические последствия аутизма и рост числа случаев РАС во всем мире указывают на острую необходимость в разработке легко реализуемых и эффективных методов скрининга. Таким образом, быстрый и доступный скрининг РАС неизбежен, чтобы помочь медицинским работникам и информировать людей о том, следует ли им проводить формальный клинический диагноз. Быстрый рост числа случаев РАС во всем мире требует наличия наборов данных, связанных с чертами поведения. Однако такие наборы данных встречаются редко, что затрудняет проведение тщательного анализа для повышения эффективности, чувствительности, специфичности и прогностической точности процесса скрининга РАС. В настоящее время доступны очень ограниченные наборы данных об аутизме, связанные с клиническими исследованиями или скринингом, и большинство из них имеют генетическую природу. Следовательно, мы предлагаем новый набор данных, связанный со скринингом взрослых на аутизм, который содержит 20 признаков, которые можно использовать для дальнейшего анализа, особенно для определения влиятельных аутичных черт и улучшения классификации случаев РАС. В этом наборе данных мы записываем десять поведенческих характеристик (AQ-10-Adult) плюс десять индивидуальных характеристик, которые доказали свою эффективность в выявлении случаев РАС среди контрольной группы в науке о поведении.

### II. Анализ данных

***Зависимость аутизма от пола:***

```python
a = pd.crosstab(data['gender'], data['Class/ASD'])
b = pd.DataFrame(a, index = ['female', 'male'], columns=[0, 1, 'sum', 'ASD Percent, %'])
b['sum'] = [a.iloc[0].sum(), a.iloc[1].sum()]
c = a.iloc[0, [1]].div(a.iloc[0].sum()).mul(100)
d = a.iloc[1, [1]].div(a.iloc[1].sum()).mul(100)
b['ASD Percent, %'] = [round(c[1], 2), round(d[1], 2)]
b
```

```python
a = pd.crosstab(data['gender'], data['Class/ASD'])
df = pd.DataFrame(a, index = ['female', 'male'], columns=[0, 1, 'sum', 'ASD Percent, %'])
b = df[0] + df[1]
df['sum'] = b
c = df[1] / df["sum"] * 100
df['ASD Percent, %'] = round(c, 2)
df
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data[['gender','Class/ASD']].groupby(['gender']).mean().plot.bar(ax=ax[0])
ax[0].set_title('Autism in subjects')
sns.countplot(x='gender', hue='Class/ASD', data=data, ax=ax[1])
ax[1].set_title('Autism distribution in each gender')
plt.show()
```

В представленной выборке женщин, страдающих аутизмом (ASD) больше, чем мужчин.  
Если проанализировать процентное содержание женщин, страдающих аутизмом, и мужчин, страдающих аутизмом, данные получаются сходными: процентное содержание женщин, страдающих аутизмом, чуть выше, чем мужчин, страдающих аутизмом. Вследствие чего можно сделать вывод, что аутизм чуть чаще встречается среди женщин, чем среди мужчин (хотя и не намного: в данной выборке разница - всего лишь около 7 %).

```python
a = pd.crosstab(data['ethnicity'], data['Class/ASD'])
df = pd.DataFrame(a, columns = [0, 1, 'sum', 'ASD Percent, %'])
b = df[0] + df[1]
df['sum'] = b
c = df[1] / df["sum"] * 100
df['ASD Percent, %'] = round(c, 2)
df.sort_values('ASD Percent, %', ascending = False)
```

По численному распределению рас можно сделать следующее заключение: большинство протестированных, страдающих аутизмом, относятся к латиноамериканцам, белым европеоидам, черным и испанцам. По процентному распределению получаются схожие результаты: среди латиноамериканцев, белых европеоидов, черных и испанцев гораздо чаще встречаются страдающие аутизмом, чем среди азиатов. Среди южных азиатов страдающие аутизмом встречаются реже всего.

При анализе зависимости аутизма от пола среди разных рас можно сделать следующее заключение: среди азиатов (а в особенности среди южноазиатов), а также среди турков аутизм чаще встречается среди мужчин. Для остальных рас, включая белых европиоидов, чернокожих и испанцев - то есть рас, среди которых, в принципе, чаще встречаются страдающие аутизмом - аутизм встречается чаще среди женщин, поэтому при изначальном анализе зависимости аутизма от пола было сделано заключение, что аутизм чаще встречается среди женщин. По текущему графику можно заключить, что, если анализировать половое распределение страдающих аутизмом по расовой принадлежности, у разных рас аутизм может встречаться чаще как у женщин, так и у мужчин.

Проанализируем зависимость аутизма от врожденного заболевания желтухой и зависимость аутизма от наличия родственников, страдающих аутизмом (более интересные признаки)

Ни один признак самостоятельно не является показателем аутизма. Только скрининг по всем десяти признакам может предсказывать аутизм при наборе от 7 и более очков по рандомным признакам.

```python
plt.figure(figsize=(8,5))
plt.rcParams['font.size'] = '10'
data.groupby('Class/ASD')['result'].plot(kind='kde')
plt.legend(['Subjects without autism', 'Subject with autism'])
plt.xlabel('AQ-scoring')
plt.show()
```

Максимальное число субъектов, страдающих аутизмом имеют скоринг 7 и 8, далее с увеличением скоринга количество субъектов идет на спад. При этом практически нет перекрытия по скорингу субъектов, страдающих и не страдающих аутизмом. Субъекты со скорингом 6 не страдают аутизмом.

***Анализ корреляции численных показателей и наличия аутизма***

```python
fig, ax = plt.subplots(figsize=(18,15))
sns.heatmap(data.corr(), annot= True, linewidths=0.2, cmap="RdYlGn", annot_kws={'size':14})
fig=plt.gcf()
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
corr = data.corr()
corr['Class/ASD'].sort_values(ascending = False)
```

Анализ корреляции показывает, что скорринговая система из десяти поведенческих характеристик (result) имеет максимальную корреляцию с наличием заболевания РАС (аутизм) у субъектов, при этом отдельно поведенческие характеристики имеют гораздо меньшую корреляцию с наличием заболевания: от 0.3 до 0.64. Индивидуальные харктеристики субъекта (возраст, наличие врожденной желтухи, наличие родственников с аутизмом) имеют слабую положительную корреляцию с наличием заболевания. Минимальная положительная корреляция у наличия заболевания с прохождением субъектом данного теста ранее. Отрицательной корреляции ни у одного из предложенных признаков нет.

### Заключение
**Разработанный скрининговый тест на аутизм, содержащий 10 поведенческих характеристик и 10 индивидуальных характеристик должен быть очень эффективен в выявлении аутизма. Данные 10 поведенчесчких характеристик по 10-балльному скоррингу позволили выявить аутизм у всех субъектов в выборке: все субъекты со скорингом 7 и более страдают аутизмом. При этом все 10 поведенческих характеристик имеют довольно схожий индивидуальный вклад в постановку диагноза.**

**Индивидуальные же характеристики, хоть и имеют положительную корреляцию с болезнью аутизмом, но значительно меньшую. При анализе индивидуальных характеристик можно заключить, что в предложенной выборке латиноамериканцы, белые европиоды, чернокожие и испанцы, чаще страдают аутизмом, чем азиаты (в особенности южные азиаты). При этом среди латиноамериканцев, белых европиодов, чернокожих и испанцец женщины чуть чаще страдают аутизмом, чем мужчины, а также большинство страдающих аутизмом имеют возраст 20 или 30 лет. Возрастная группа в выборке представлена недостаточно, поэтому наиболее часто встречающийся возраст среди субъектов с аутизмом может быть представлен неверно. Наличие врожденной желтухи и родственников, страдающих аутизмом имеет положительное влияние на развитие аутизма: среди субъектов с врожденной желтухой и с родственниками, страдающими аутизмом, чаще встречаются субъекты с аутизмом.**
