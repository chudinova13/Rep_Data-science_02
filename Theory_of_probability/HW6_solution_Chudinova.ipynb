### Задача 1

Вы открываете бутылку молока и делаете из нее глоток. Вероятность того, что молоко скисло - 0.02. С какой вероятностью молоко окажется свежим?

Решение.  
A: молоко скисло. P(A) = 0.02  
B: молоко не скисло. P(B) = 1 - P(A) = 0.98

**Ответ:** молоко окажется свежим с вероятностью 0.98

### Задача 2

На пешеходном переходе установлен светофор с 2 лампами. Вероятность перегорания лампы в течение года - 0.2. С какой вероятностью не перегорит хотя бы одна лампа?

Решение.  
A: первая лампа перегорит в течение года. P(A) = 0.2  
B: вторая лампа перегорит в течение года. P(B) = 0.2  
C: обе лампы перегорели: P(C) = P(A) * P(B) = 0.04  
не С (хотя бы одна лампа не перегорела) = 1 - 0.04 = 0.96  

Проверяю:  
A: Не перегорит первая лампа, вторая перегорит  
B: Перегорит первая лампа, вторая не перегорит  
C: Не перегорят обе лампы  
P(A) + P(B) + P(C) = (1 - 0.2) * 0.2 + 0.2 * (1 - 0.2) + (1 - 0.2) * (1 - 0.2) = 0.16 + 0.16 + 0.64 = 0.96

**Ответ:** хотя бы одна лампа не перегорит с вероятностью 0.96

### Задача 3

Менеджер по продажам совершает сделку если знаком с клиентом с вероятностью - 0.8, а если ничего о нем не слышал с вероятностью 0.1. Начальник передает менеджеру список из 10 клиентов, 4 из которых менеджеру известны. С какой вероятностью выбрав случайного клиента из списка менеджер ничего ему не продаст?

Решение.  
Клиент знаком: 0.8 - совершает сделку, 0.2 - не совершает сделку  
Клиент незнаком: 0.1 - совершает сделку, 0.9 - не совершает сделку  

доля знакомых клиентов: 4/10 = 0,4  
доля незнакомых клиентов: 6/10 = 0,6  

А: не совершается сделка со знакомым из списка: P(A) = 0.4 * 0.2 = 0.08  
B: не совершается сделка с незнакомым из списка: P(B) = 0.6 * 0.9 = 0.54  
P(A + B) = 0.08 + 0.54 = 0.62

**Ответ:** ничего не будет продано случайному клиенту из списка с вероятностью 0.62
